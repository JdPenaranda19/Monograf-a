```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
archivos = [1,3,5,7,10]
filename = 'C:/Users/Jose peñaranda/Documents/Documentos python/env/archivos/Archivos_Suavizados/velocity_AbacusCosmos_720box_planck_00_0_FoF_vmax_300.0_sigma_{}.0_nside_360.hdf5'

divergencias = []


for i in archivos:
    with h5py.File(filename.format(i), 'r') as f:
        print(f.keys())
        divergence = list(f.keys())[0]

        dataDivergence = np.array(f[divergence])
    divergencias.append(dataDivergence)
    

std = []
for i in divergencias:
    std.append(np.std(i.flatten()))
```

```python
plt.figure(figsize = (6.5,6.55))
plt.rcParams.update({'font.size': 16})
plt.semilogy(archivos*2,std, '--')
plt.semilogy(archivos*2,std,'o',c = 'black')
plt.xlabel("$\sigma_{s}$(Mpc/h)", fontsize = 18)
plt.ylabel(r'Std of the divergence[$\frac{km\cdot h}{Mpc\cdot s}]$', fontsize = 16)
plt.savefig("std_smooth.png")
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
archivos = [3,5,7,10]
filename = 'C:/Users/Jose peñaranda/Documents/Documentos python/env/archivos/Archivos_Suavizados/velocity_AbacusCosmos_720box_planck_00_0_FoF_vmax_300.0_sigma_{}.0_nside_360.hdf5'

velocidades = []
velocidades2 = []

divergencias = []
datos_seleccionados = []
datos_seleccionados2 = []

aleatorio = 250
for i in archivos:
    with h5py.File(filename.format(i), 'r') as f:
        print(f.keys())
        divergence = list(f.keys())[0]
        vel_x = list(f.keys())[1]
        vel_y = list(f.keys())[2]
        vel_z = list(f.keys())[3]

        dataDivergence = np.array(f[divergence])
        data_vel_x = np.array(f[vel_x])
        data_vel_y = np.array(f[vel_y])
        data_vel_z = np.array(f[vel_z])
    velocidades.append(np.sqrt(data_vel_x**2 + data_vel_y**2 + data_vel_z**2))
    divergencias.append(dataDivergence)
```

```python
rangos = [(-5,5),(-3,3),(-1,1),(-0.5,0.5)]
plt.figure(figsize = (12,12))
plt.rcParams.update({'font.size': 28})
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.hist(divergencias[i].flatten(), bins = 50, density = True,edgecolor = 'black', range = rangos[i],histtype = 'step',lw = 2)
    plt.title(r"$\sigma_s $ = {}Mpc/h".format(archivos[i]*2), fontsize = 28)
    if(i>=2):
        plt.xlabel(r'Divergence[$\frac{km\cdot h}{Mpc\cdot s}]$', fontsize = 28)
    if(i%2 == 0):
            plt.ylabel("PDF", fontsize = 28)
plt.subplots_adjust(hspace=0.3, wspace=0.3)
plt.savefig("smooth_grad_dist.png")
```

```python
rangos = [(0,40),(0,25),None,None]
plt.figure(figsize = (12,12))
plt.rcParams.update({'font.size': 26})
for i in range(4):
    vel_actual = []
    plt.subplot(2,2,i+1)
    plt.hist(velocidades[i].flatten(), bins = 50, density = True,edgecolor = 'black', range = rangos[i],histtype = 'step', lw = 2)
    plt.title(r"$\sigma_s $ = {}Mpc/h".format(archivos[i]*2), fontsize = 26)
    if(i>=2):
        plt.xlabel("Speed[km/s]", fontsize = 26)
    if(i%2 == 0):
        plt.ylabel("PDF", fontsize = 26)
    
plt.subplots_adjust(hspace=0.3, wspace=0.25)

plt.savefig("smooth_vel_dist.png")
```
